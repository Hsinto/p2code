```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Partially decoupled system

def diffeq_x(x, y):
    x_ = 9*x - 3*x**2 - 3*x*y
    return(x_)
def diffeq_y(x, y):
    y_ = -2*y + x*y
    return(y_)
```

```python
def g(x_not, y_not, range_t, iterations):
    delta = range_t/iterations
    x_array = []
    y_array = []
    x_next = x_not
    y_next = y_not

    for i in range(0, iterations):
        x_array.append(x_next)
        y_array.append(y_next)
        y_next = y_next + delta*diffeq_y(x_next, y_next)
        x_next = x_next + delta*diffeq_x(x_next, y_next)


    # print(x_array)
    # print(y_array)
    
    # Graph functions

    x=x_array
    y=y_array

    # plt.plot(x, y, marker = ".", markersize = 3)
    # plt.xlim(-2, 2)
    # plt.ylim(-2, 2)
    # plt.show()
    return x, y
```

```python
def plot_diff(*myplots):
  for plot in myplots:  # Takes each plot in the tuple myplots
    result = g(*plot) # Calls the function to find x,y
    x = result[0]  # Accesses the x values
    y = result[1]  # Accesses the y values
    # x_not = x[0]
    # y_not = y[0]
    plt.plot(x,y, 
            #  marker = '.', markersize = 8, label = f"x(0) = {x[0]}, y(0) = {y[0]}"
            )
```

```python
plot_diff(
        # [0,0,10,1000],
        # [5,2,10,1000],
        [0,4,10,1000],
        # [2,3,10,1000],
        # [4,2,10,1000],
        # [6,4,10,1000],
        [2,0,10,1000],
        [5,0,10,1000]
        
        )

# plt.legend()
plt.xlim(-1,7)
plt.ylim(-1,7)

plt.show()
```

```python
def diffeq_1t(x_not, y_not, range_t, iterations):
    delta = range_t/iterations
    x_array = []
    y_array = []
    t_array = []
    x_next = x_not
    y_next = y_not
    t_next = 0

    for i in range(0, iterations):
        x_array.append(x_next)
        y_array.append(y_next)
        t_array.append(t_next)
        y_next = y_next + delta*diffeq_y(x_next, y_next)
        x_next = x_next + delta*diffeq_x(x_next, y_next)
        t_next = t_next + delta


    x=x_array
    y=y_array
    t=t_array

    return x, y, t
    
```

```python
def plot_relation(*myplots):
  for plot in myplots:  # Takes each plot in the tuple myplots
    result = diffeq_1t(*plot) # Calls the function to find x,y
    x = result[0]  # Accesses the x values
    y = result[1]  # Accesses the y values
    t = result[2]
    # x_not = x[0]
    # y_not = y[0]
    plt.plot(t,x, label = "x(t)")
    plt.plot(t,y, label = "y(t)")
```

```python
plot_relation(
        [0, 4, 10, 1000],
        [2,0,10,1000]
)

plt.legend()
plt.xlim(0,6)
plt.ylim(0,5)

plt.show()
```
